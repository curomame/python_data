```python
# Rolling Expanding
```

```python
# 기본 전제는 시계열 데이터의 일반적 프로세스는 rolling mean을 기반으로 하는 데이터임
# 그래서 데이터를 시간의 창으로 분할하는 작업을 진행해줘야함
# 그리고 각 창에 대해 집계 함수를 계산

# 이 방식으로 우리는 단순 이동 평균을 구할 수 있음
```

```python
import pandas as pd
```

```python
%matplotlib inline
```

```python
df= pd.read_csv('/Users/leehyeongmin/python_data/lecture/Data/starbucks.csv',index_col='Date',parse_dates=True)
```

```python
df.head()
```

```python
df['Close'].plot(figsize=(12,5))
```

```python
df['Close'].plot(figsize=(12,5))
df.rolling(window=60).mean()['Close'].plot()
```

```python
df['Close : 30 Day Mean'] = df['Close'].rolling(window=30).mean()
```

```python
df.head()
```

```python
df[['Close','Close : 30 Day Mean']].plot(figsize=(12,5))
```

```python
df['Close'].expanding().mean().plot(figsize=(12,5))
```
